```python
import rebound
import numpy as np
import matplotlib.pyplot as plt
from astropy import units as u

def to_radians(value):
    return (value*u.deg).to(u.rad).value

sim = rebound.Simulation()
sim.units = ('yr', 'AU', 'Msun')
sim.add(m=1.0)
sim.add(m=1.0/1000/300, a=1.0, omega=0, Omega=0, e=0, inc=to_radians(100))
sim.add(m=1.0/1000/300, a=1.0, omega=0, Omega=0, e=0, inc=to_radians(0))

plt.figure()
fig = rebound.OrbitPlot(sim, unitlabel="[AU]", periastron=True, lw=2)
plt.title('Simple')
plt.show()
```

```python
sim = rebound.Simulation()
sim.units = ('yr', 'AU', 'Msun')
sim.add(m=1.0)
sim.add(m=1.0/1000, a=1.0, omega=0, Omega=0, e=0, inc=to_radians(100))
sim.add(m=1.0/1000, a=1.0, omega=0, Omega=0, e=0.1, inc=to_radians(0))

step_in_time=1000

sim.integrate(step_in_time)

plt.figure()
fig = rebound.OrbitPlot(sim, unitlabel="[AU]", periastron=True, lw=2)
plt.title('Simple')
plt.show()
```
